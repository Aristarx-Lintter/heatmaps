```python
import torch
from transformers import AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info

messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": "https://qianwen-res.oss-cn-beijing.aliyuncs.com/Qwen-VL/assets/demo.jpeg",
            },
            {"type": "text", "text": "Describe this image."},
        ],
    }
]

# Preparation for inference
text = processor.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt",
)
```

```python
def gen_inputs(inputs, max_new_tokens=1000):
    generated_ids = local_model.generate(**inputs, max_new_tokens=max_new_tokens)
    generated_ids_trimmed = [
        out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    output_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    return output_text, generated_ids, generated_ids_trimmed

with torch.no_grad():
    output_text, generated_ids_heat, _ = gen_inputs(inputs.to("cuda"))
print(output_text)
```

```python
from src.common.templates import messages_template, answer_template
from src.common.dataset import process_injection


def make_pure_inputs(part):
    messages = messages_template(part["image"], part['upd_transcribation'])[:1]
    text = processor.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )
    image_inputs, video_inputs = process_vision_info(messages)
    inputs = processor(
        text=[text],
        images=image_inputs,
        videos=video_inputs,
        padding=True,
        return_tensors="pt",
    )
    return inputs

def add_heatmaps(inputs, part):
    heat_inputs = inputs.copy()
    heat_inputs["heatmap_flat"] = process_injection(inputs["image_grid_thw"], [part]).to("cuda")  # 
    heat_inputs = heat_inputs.to("cuda")

    return heat_inputs
```

```python
from PIL import Image, ImageOps
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

def create_pil_heatmap(pil_image: Image.Image) -> Image.Image:
    width, height = pil_image.size
    heatmap_np = np.zeros((height, width), dtype=np.uint8)
    gradient = np.linspace(0, 255, width, dtype=np.uint8)
    heatmap_np = np.tile(gradient, (height, 1))
    heatmap_pil = Image.fromarray(heatmap_np, mode='L')
    return heatmap_pil

def draw_heatmap_on_image_from_path(image_path: str, heatmap_pil: Image.Image, alpha: float = 0.5, colormap: int = cv2.COLORMAP_JET) -> np.ndarray:
    image_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    heatmap_gray = np.array(heatmap_pil)
    
    target_height, target_width = image_bgr.shape[:2]
    heatmap_gray = cv2.resize(heatmap_gray, (target_width, target_height), interpolation=cv2.INTER_LINEAR)
    
    heatmap_normalized = cv2.normalize(heatmap_gray, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    heatmap_color = cv2.applyColorMap(heatmap_normalized, colormap)
    overlay = cv2.addWeighted(heatmap_color, alpha, image_bgr, 1.0 - alpha, 0.0)
    return overlay

def draw_and_print_answer(part):
    with torch.no_grad():
        inputs = make_pure_inputs(part)
        output_text, generated_ids_heat, _ = gen_inputs(inputs.to("cuda"))
        print("Vanilla output:", output_text[0])
        print("*"*20, "\n")
        heat_inputs = add_heatmaps(inputs, part)
        output_text, generated_ids_heat, _ = gen_inputs(heat_inputs)
        print("Heatmap injected:", output_text[0])
        print("*"*20, "\n")
        print("Target:", part['upd_transcribation'])
        print("*"*20, "\n\n\n")

    result_overlay_bgr = draw_heatmap_on_image_from_path("./images/"+part['image'], part['heatmap'], alpha=0.6, colormap=cv2.COLORMAP_VIRIDIS)
    
    result_overlay_rgb = cv2.cvtColor(result_overlay_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 5))
    plt.imshow(result_overlay_rgb)
    plt.title('Image with Overlayed Heatmap')
    plt.axis('off')
    plt.show()
```

```python
def create_corner_heatmap(pil_image: Image.Image, corner_fraction: float = 0.45) -> Image.Image:
    height, width = image_bgr.shape[:2]
    heatmap_np = np.zeros((height, width), dtype=np.uint8)

    corner_h = int(height * corner_fraction)
    corner_w = int(width * corner_fraction)
    heatmap_np[20:(corner_h-50), 100:(corner_w-80)] = 255

    heatmap_pil = Image.fromarray(heatmap_np, mode='L')
    return heatmap_pil

def draw_heatmap_on_image_from_image(image_path: str, heatmap_pil, alpha: float = 0.5, colormap: int = cv2.COLORMAP_JET) -> np.ndarray:
    image_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
    
    target_height, target_width = np.array(heatmap_pil).shape[:2]
    image_bgr = cv2.resize(image_bgr, (target_width, target_height), interpolation=cv2.INTER_LINEAR)
    new_heat = create_corner_heatmap(image_bgr)
    new_heat = np.array(new_heat)

    print(new_heat.shape, image_bgr.shape)
    
    heatmap_normalized = cv2.normalize(new_heat, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    heatmap_color = cv2.applyColorMap(heatmap_normalized, colormap)
    overlay = cv2.addWeighted(heatmap_color, alpha, image_bgr, 1.0 - alpha, 0.0)
    return overlay

Image.fromarray(draw_heatmap_on_image_from_image("./images/"+part['image'], part['heatmap']))
```
